**С последними же шестью переменными в таблице мы вряд ли сможем что-то сделать, поэтому оставим пропуски.**



Разберемся с переменными living_area_m2  (жилая площадь в квадратных метрах) и kitchen_area_m2 (площадь кухни в квадратных метрах):

```python
# Посмотрим на корреляционную матрицу, чтобы понять, с какими переменными больше всего коррелирует переменная kitchen_area_m2
df.corr().style.background_gradient('coolwarm')
```

```python
# создадим коэффициенты, чтобы уточнить данные, разделив площади (жилые и кухни) на общую 
df['liv_coef'] = df['living_area_m2']/df['total_area_m2'] 

df['kit_coef'] = df['kitchen_area_m2']/df['total_area_m2']
```

```python
df.head()
```

```python
# сгруппируем по количеству комнат, запишем медианный коэффициент в словарь
liv_coef_dict = df.groupby('number_of_rooms')['liv_coef'].median().to_dict() 

kit_coef_dict = df.groupby('number_of_rooms')['kit_coef'].median().to_dict()
```

```python
# воспользуемся методом map, который  меняет на значение из словаря в зависимости от значения в ключевом столбце
df['liv_coef'] = (df['number_of_rooms'].map(liv_coef_dict))*df['total_area_m2']

df['kit_coef'] = (df['number_of_rooms'].map(kit_coef_dict))*df['total_area_m2'] 
```

```python
# заменим пропущенные значения в переменных living_area_m2 и kitchen_area_m2
df['living_area_m2'] = df['living_area_m2'].fillna(df['liv_coef'])

df['kitchen_area_m2'] = df['kitchen_area_m2'].fillna(df['kit_coef'])
```

```python
# Сгруппируем данные по количеству комнат и посмотрим на медианные значения площади кухни в квадратных метрах в каждой группе
df.groupby('number_of_rooms')['kitchen_area_m2'].median()
```

Первая строчка таблицы говорит о том, что отсутствующие значения по данному показателю принадлежат квартирам с нулевым числом комнат, что достаточно логично)

Есть предположение, что дубликаты могут возникнуть, если привести названия населенных пунктов к более общему виду. Проверим это:

```python
# Напишем функцию, которая поможет нам сократить количество теста и избавиться от таких слов, как 'поселок','деревня':
def no_upper(string): 
    '''
    Проверяем каждый символ строки на большую букву.
    Как только большая буква попадается - возвращаем остаток строки от большой буквы включительно до последнего символа
    '''
    for i in range(len(string)):
        if string[i].isupper():
            return string[i: len(string)]
```

```python
df['town_name'] = df['town_name'].apply(no_upper) 
```

```python
df['town_name'].unique()
```

```python
len(df['town_name'].unique())
```

### Исключение редких и выбивающихся значений

```python
# Напишем функцию, которая будет устранять выбросы
def delete_outliers(df, column):
    q1 = df[column].quantile(0.25)                 
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    cutoff = (df[column] >= (q1 - 1.5*iqr)) & (df[column] <= (q3 + 1.5*iqr))
    return df.loc[cutoff]
```

Для начала разберемся с теми переменными, которые мы изучали в прошлом разделе. А именно с ценой, площадью, количеством комнат и высотой потолков
